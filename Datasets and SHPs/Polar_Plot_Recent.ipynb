```python
def hex_to_rgb(value):
    '''
    Converts hex to rgb colours
    value: string of 6 characters representing a hex colour.
    Returns: list length 3 of RGB values'''
    value = value.strip("#") # removes hash symbol if present
    lv = len(value)
    return tuple(int(value[i:i + lv // 3], 16) for i in range(0, lv, lv // 3))


def rgb_to_dec(value):
    '''
    Converts rgb to decimal colours (i.e. divides each value by 256)
    value: list (length 3) of RGB values
    Returns: list (length 3) of decimal values'''
    return [v/256 for v in value]
```

```python
import matplotlib.colors as mcolors

def get_continuous_cmap(hex_list, float_list=None):
    ''' creates and returns a color map that can be used in heat map figures.
        If float_list is not provided, colour map graduates linearly between each color in hex_list.
        If float_list is provided, each color in hex_list is mapped to the respective location in float_list. 
        
        Parameters
        ----------
        hex_list: list of hex code strings
        float_list: list of floats between 0 and 1, same length as hex_list. Must start with 0 and end with 1.
        
        Returns
        ----------
        colour map'''
    rgb_list = [rgb_to_dec(hex_to_rgb(i)) for i in hex_list]
    if float_list:
        pass
    else:
        float_list = list(np.linspace(0,1,len(rgb_list)))
        
    cdict = dict()
    for num, col in enumerate(['red', 'green', 'blue']):
        col_list = [[float_list[i], rgb_list[i][num], rgb_list[i][num]] for i in range(len(float_list))]
        cdict[col] = col_list
    cmp = mcolors.LinearSegmentedColormap('my_cmp', segmentdata=cdict, N=256)
    return cmp
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm


plt.rcParams.update({'font.size': 50})
import math


N = len(a)
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
r = np.array(a['Percent_pop_covered'])

label = list(a.index.values)

width = np.array([0.15] * N)

ax = plt.subplot(111, projection='polar')
ax.set_rlim(0, 300)
ax.set_rticks(np.arange(0, 101, 20)) #lines
ax.set_thetagrids(theta * 180 / np.pi)
ax.set_yticklabels([]) #position of grid labels
ax.xaxis.grid(False)
ax.yaxis.grid(True, color = '#195D7E', linestyle = '-' )
ax.set_facecolor("#F5F5EF")

#colors = plt.cm.hsv(theta/5/np.pi)


data_color = a['GDP p/capita']


data_color = np.array([x / max(data_color) for x in data_color])

my_cmap = plt.cm.get_cmap(get_continuous_cmap(['#F17070', '#195D7E','#05997A']))
colors = my_cmap(data_color)
ax.spines['polar'].set_visible(False)


ax.bar(x=theta, height=r, width=width, bottom=0, alpha=1, tick_label='', color=colors) #tick_label=label,

from matplotlib.cm import ScalarMappable


sm = ScalarMappable(cmap=my_cmap, norm = mcolors.TwoSlopeNorm(vmin=min(GDP),vcenter=np.median(GDP), vmax=max(GDP)))
sm.set_array([])

cbar = plt.colorbar(sm,)
cbar.set_label('GDP per capita (â‚¬)',labelpad=-450, weight='bold', fontsize = 70)
cbar.ax.tick_params(labelsize=70)

plt.gcf().set_size_inches(50, 50)

plt.savefig('polar_plot.png', transparent=True, dpi=200)


plt.show()
```
